# Squared Continuous

```python
def exp_f(base, x):
    return (base ** x - 1) / (base - 1)

x = np.linspace(0, 1, 100)
linear_y = x
exp_y = np.array(list(map(lambda e: exp_f(10, e), x)))
sq_y = x ** 2 # Winner
tr_y = x **3
```

```python
plt.plot(x, linear_y)
plt.plot(x, exp_y)
plt.plot(x, sq_y)
plt.plot(x, tr_y)
```
